```python
tips = sns.load_dataset("tips") # small dataset
titanic = sns.load_dataset("titanic") # medium dataset
diamonds = sns.load_dataset("diamonds") # large dataset
```

### Most Popular **histplot()** Parameters (sorted by use-case popularity)

1.	**data** = 	DataFrame or array containing the data
2.	**x**	= Column name or array for x-axis variable
3.	**y**	= Column name or array for y-axis variable (2D histogram)
4.	**bins**	= Number of bins (int or sequence)
5.	**binwidth** = 	Width of each bin
6.	**binrange** =	Tuple (min, max) to set bin range
7.	**kde**	= Show kernel density estimate (True/False)
8.	**hue**	= Grouping variable (e.g., sex, cut)
9.	**multiple**	= How to display multiple histograms: layer, stack, dodge, fill
10.	**element**	= Bar style: bars, step, poly
11.	**stat**	= Y-axis: count, frequency, density, probability
12.	**common_norm**	= Normalize across groups (True or False)
13.	**discrete**	= Treat variable as discrete (used with integers/categoricals)
14.	**log_scale**	= Apply log scale to axes (True, or (True, False))
15.	**cbar**	= Add color bar (used for 2D histograms)
16.	**palette**	= Color palette for hue categories
17.	**color**	= Single color for bars
18.	**alpha**	= Transparency (0 to 1)
19.	**fill**	= Fill bars (default True; set False for outline only)
20.	**line_kws**	= Keyword arguments for KDE line style (dict)
21.	**ax**	= Matplotlib axes object to plot on
22.	**weights**	= Array of weights to apply to values
23.	**hue_order**	= Order in which to draw hue levels
24.	**legend**	= Show/hide legend (True/False or "auto")
25.	**orientation**	= v or h (vertical/horizontal)
26.	**fill_kws**	= Styling for filled area of KDE (dict)

### Working small data-sets

```python
sns.histplot(data=tips,x="total_bill" ,bins=10, kde=True)
sns.histplot(data=tips,x="total_bill" ,bins=20, kde=True)
```

```python
sns.histplot(data=tips,x="total_bill" ,bins=10, binwidth=10, kde=True)
sns.histplot(data=tips,x="total_bill" ,bins=10,binwidth=5, kde=True)
```

```python
sns.histplot(data=tips,x="total_bill" ,bins=10,binwidth=5,binrange=(30,60))
# binrange is used to limit the range of the bins 
```

```python
sns.histplot(data=tips,x="total_bill" ,bins=10, binwidth=5, kde=True, hue="smoker")
# hue is used to color the bars based on a categorical variable
```

####  must use hue=... when using multiple!

```python
# STACKED HISTOGRAM
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=10,
    binwidth=5,
    kde=True,
    multiple="stack"
)
```

```python
# DODGED (SIDE-BY-SIDE) HISTOGRAM
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=10,
    binwidth=5,
    kde=True,
    multiple="dodge"
)
```

```python
# FILLED HISTOGRAM (100% PER BIN)
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=10,
    binwidth=5,
    kde=True,
    multiple="fill"
)
```

#### The stat parameter controls what the histogram shows on the y-axis.

```python
sns.histplot(data=tips, x="total_bill", stat="count")
# The stat parameter controls what the histogram shows on the y-axis.
# count: number of observations in each bin
```

```python
sns.histplot(data=tips, x="total_bill", stat="frequency")
# frequency: number of observations in each bin divided by the total number of observations
```

```python
sns.histplot(data=tips, x="total_bill", stat="probability")
# probability: number of observations in each bin divided by the total number of observations, normalized to sum to 1
```

```python
sns.histplot(data=tips, x="total_bill", stat="density", kde=True)
# density: number of observations in each bin divided by the total number of observations and the bin width, normalized to sum to 1
```

```python
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    stat="density",
    common_norm=False,
    kde=True,
    multiple="layer"
)
# This allows for separate normalization of each hue category, useful for comparing distributions across categories.
# The `common_norm=False` parameter ensures that the densities for each hue category are not combined, allowing for a clearer comparison of their individual distributions.
```
