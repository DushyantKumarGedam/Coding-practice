### seaborn parameter
1. data = specific dataset 
2. x = x variable for axes
3. y = y variable for axes
4. hue = different in varable colour
5. size = width of element
6. style = style of element
7. palette = colour scheme for **hue**


## typies of plot in seaborn :-
### relplot(relational)
1. scatterplot
2. lineplot

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
```

### displot(distributions)
1. histplot
2. kdeplot
3. ecdfplot
4. rugplot 

```python
sns.histplot(tips["total_bill"])
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.ecdfplot(tips["total_bill"])
```

```python
sns.rugplot(tips["total_bill"])
```

### catplot(categorical)
1. stripplot
2. swarmplot
3. boxplot
4. violinplot
5. pointplot
6. barplot

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.pointplot(x="day", y="total_bill", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
```

### Other plots
1. Headmap
2. Pairplot
3. Joint plot
4. facet grid

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Compute correlation only on numeric columns
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.pairplot(data=tips)
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="scatter")
plt.show()
```

```python
g= sns.FacetGrid(tips, col="day")
g.map(sns.histplot, "total_bill")
```
